```python

# №3 Створіть масив розміром 3x3, заповніть його випадковими дійсними числами від 0 до 1 та знайдіть мін. елемент.
import numpy as np

matrix = np.random.rand(3, 3)

min_element = np.min(matrix)

print("Масив розміром 3x3 з випадковими дійсними числами від 0 до 1:")
print(matrix)
print("\nМінімальний елемент у масиві:")
print(min_element)
```

```python
# №4 Створіть масив розміром 3x3, заповніть його випадковими дійсними числами в діапазоні від 0 до 1 та знайдіть вектор-стовпчик, що містить мінімальні елементи в кожному рядку.
import numpy as np

matrix = np.random.rand(3, 3)

min_vector = np.min(matrix, axis=1).reshape(-1, 1)

print("Масив розміром 3x3 з випадковими дійсними числами від 0 до 1:")
print(matrix)
print("\nВектор-стовпчик, що містить мінімальні елементи в кожному рядку:")
print(min_vector)
```

```python
# №5 Створіть два одновимірних масиви розміром 4, заповніть їх випадковими цілими числами від -5 до 5 та виконайте на них поелементні операції додавання, віднімання та множення.
import numpy as np

array1 = np.random.randint(-5, 6, 4)
array2 = np.random.randint(-5, 6, 4)

addition = array1 + array2
subtraction = array1 - array2
multiplication = array1 * array2

print("Перший масив:")
print(array1)
print("\nДругий масив:")
print(array2)

print("\nРезультат поелементного додавання:")
print(addition)
print("\nРезультат поелементного віднімання:")
print(subtraction)
print("\nРезультат поелементного множення:")
print(multiplication)
```

```python
# №6 Створіть дві матриці 3x3, заповніть їх довільними цілими числами від 1 до 10 та знайдіть їх добуток
import numpy as np

matrix1 = np.random.randint(1, 11, (3, 3))
matrix2 = np.random.randint(1, 11, (3, 3))

product = np.dot(matrix1, matrix2)

print("Перша матриця:")
print(matrix1)
print("\nДруга матриця:")
print(matrix2)
print("\nДобуток двох матриць:")
print(product)
```

```python
# №7 Створіть дві матриці розміром 2x3 та 3x3, заповніть їх випадковими цілими числами в діапазоні від 1 до 4 та перемножте їх між собою.
import numpy as np

matrix1 = np.random.randint(1, 5, (2, 3))

matrix2 = np.random.randint(1, 5, (3, 3))

product = np.dot(matrix1, matrix2)

print("Перша матриця (2x3):")
print(matrix1)
print("\nДруга матриця (3x3):")
print(matrix2)
print("\nДобуток двох матриць (2x3 * 3x3):")
print(product)
```

```python
# №8 Створіть матрицю розміром 3x3, заповніть її випадковими цілими числами в діапазоні від 1 до 12 та знайдіть її обернену матрицю.
import numpy as np

np.random.seed(0)

matrix = np.random.randint(1, 13, size=(3, 3))
print("Оригінальна матриця:")
print(matrix)
if np.linalg.det(matrix) != 0:
    inverse_matrix = np.linalg.inv(matrix)
    print("Обернена матриця:")
    print(inverse_matrix)
else:
    print("Матриця є виродженою і обернену матрицю знайти неможливо.")
```

```python
# №9 Створіть матрицю розміром 4x4, заповніть її випадковими дійсними числами в діапазоні від -7 до 7 та транспонуйте її.
import numpy as np

np.random.seed(0)

matrix = np.random.uniform(-7, 7, size=(4, 4))
print("Оригінальна матриця:")
print(matrix)
transposed_matrix = matrix.T
print("Транспонована матриця:")
print(transposed_matrix)
```

```python
# №10 Створіть матрицю розміром 3x4 та вектор розміром 4x1, заповніть матрицю випадковими дійсними числами від 0 до 1, 
# а вектор - випадковими дійсними числами від 0 до 1, і знайдіть добуток матриці на вектор
import numpy as np

np.random.seed(0)

matrix = np.random.uniform(0, 1, size=(3, 4))
print("Матриця 3x4:")
print(matrix)

vector = np.random.uniform(0, 1, size=(4, 1))
print("Вектор 4x1:")
print(vector)
product = np.dot(matrix, vector)
print("Добуток матриці на вектор:")
print(product)
```

```python
# №12 Створіть дві матриці розміром 9x9, заповніть їх випадковими цілими числами в діапазоні від 1 до 10 та 
# знайдіть відношення елементів другого стовпця другої матриці до відповідних елементів першої матриці (поелементне ділення).
import numpy as np

np.random.seed(0)

matrix1 = np.random.randint(1, 11, size=(9, 9))
matrix2 = np.random.randint(1, 11, size=(9, 9))

print("Перша матриця 9x9:")
print(matrix1)

print("Друга матриця 9x9:")
print(matrix2)
second_column_matrix2 = matrix2[:, 1]


result = second_column_matrix2 / matrix1[:, 1]

print("Відношення елементів другого стовпця другої матриці до відповідних елементів першої матриці:")
print(result)
```

```python
# №13 Створіть дві матриці розміром 3x3, заповніть їх випадковими цілими числами в діапазоні від 1 до 10 та знайдіть добуток їх елементів.
import numpy as np

np.random.seed(0)

matrix1 = np.random.randint(1, 11, size=(3, 3))
matrix2 = np.random.randint(1, 11, size=(3, 3))

print("Перша матриця 3x3:")
print(matrix1)

print("Друга матриця 3x3:")
print(matrix2)
product = np.dot(matrix1, matrix2)
print("Добуток матриць:")
print(product)
```

```python
# №14 Створіть матрицю розміром 3x3, заповніть її випадковими цілими числами в діапазоні від 1 до 5 та знайдіть середнє геометричне значення елементів матриці.
import numpy as np
from scipy.stats import gmean

np.random.seed(0)

matrix = np.random.randint(1, 6, size=(3, 3))
print("Матриця 3x3:")
print(matrix)

elements = matrix.flatten()

geom_mean = gmean(elements)
print("Середнє геометричне значення елементів матриці:")
print(geom_mean)
```

```python
# №15 Створіть матрицю розміром 3x3, заповніть її випадковими дійсними числами в діапазоні від -2 до 2 та знайдіть вектор-стовпчик, 
# що містить найменше за модулем ціле значення в кожному рядку матриці.
import numpy as np

matrix = np.random.uniform(-2, 2, (3, 3))

sums = np.sum(np.abs(matrix.astype(int)), axis=0)
min_column_index = np.argmin(sums)

result_column = matrix[:, min_column_index]

print("Матриця:")
print(matrix)
print("\nВектор-стовпчик:")
print(result_column.reshape(-1, 1))
```
